```python
import re

# Regex pattern to match most common emojis
emoji_pattern = re.compile(
    "[" 
    "\U0001F600-\U0001F64F"  # emoticons
    "\U0001F300-\U0001F5FF"  # symbols & pictographs
    "\U0001F680-\U0001F6FF"  # transport & map symbols
    "\U0001F1E0-\U0001F1FF"  # flags
    "\U00002702-\U000027B0"  # dingbats
    "\U000024C2-\U0001F251"  
    "]+", flags=re.UNICODE)

# Remove emojis from 'Review Content' column, keep other text unchanged
CBE['Review Content'] = CBE['Review Content'].apply(
    lambda x: emoji_pattern.sub(r'', x) if isinstance(x, str) else x
)
```
